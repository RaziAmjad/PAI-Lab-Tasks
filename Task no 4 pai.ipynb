```python
def is_safe(board, row, col, n):
    # Check this column
    for i in range(row):
        if board[i][col] == 1:
            return False

    # Check upper left diagonal
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False

    # Check upper right diagonal
    for i, j in zip(range(row, -1, -1), range(col, n)):
        if board[i][j] == 1:
            return False

    return True

def solve_n_queens_all(board, row, n, solution_count):
    # Base case: Agar row == n, matlab ek solution mil gaya
    if row == n:
        solution_count[0] += 1
        print(f"\nSolution #{solution_count[0]}:")
        print_board(board)
        return # Yahan se wapis jaayenge taake aur solutions dhund sakein

    for col in range(n):
        if is_safe(board, row, col, n):
            # Queen rakhi
            board[row][col] = 1

            # Agli row par gaye
            solve_n_queens_all(board, row + 1, n, solution_count)

            # BACKTRACK: Queen hata di taake doosri positions try ho sakein
            board[row][col] = 0

def print_board(board):
    for row in board:
        print(" ".join("Q" if x == 1 else "." for x in row))
    print("-" * (len(board) * 2))

# Dynamic Input
N = int(input("Enter the number of Queens (N): "))
board = [[0 for _ in range(N)] for _ in range(N)]
solution_count = [0] # List use ki hai taake function ke andar update ho sake

solve_n_queens_all(board, 0, N, solution_count)

if solution_count[0] == 0:
    print(f"No solution exists for N={N}")
else:
    print(f"\nTotal solutions found: {solution_count[0]}")
```
